```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing(transparent=True) as d:
    elm.style(elm.STYLE_IEC)
    bjt = elm.BjtNpn(leads=True, circle=True).label('Q1')
    elm.Line().right(d.unit/2).at(bjt.collector).dot().label('$U_{CE}$', loc='right')
    Rc = elm.Resistor().at(bjt.collector).up().label('$R_{C}$')
    elm.Line().at(Rc.end).up(d.unit/8).dot().label('$U_{CC}$', loc='top')
    baseline = elm.Line().at(bjt.base).left(d.unit/4).dot()
    elm.Line().at(baseline.end).up(d.unit/8)
    R1 = elm.Resistor().at(baseline.end).up().label('$R_{1}$')
    elm.Line().at(R1.end).up(d.unit/8)
    elm.Line().right(d.unit/2).dot()
    Re = elm.Resistor().at(bjt.emitter).down().label('$R_{E}$')
    elm.Line().at(Re.end).down(d.unit/8)
    elm.Ground().at(Re.end)
    R2 = elm.Resistor().at(baseline.end).down().label('$R_{2}$')
    elm.Line().at(R2.end).down(d.unit/8)
    elm.Line().right(d.unit/2).dot()
    elm.Line().at(baseline.end).left(d.unit/2).dot().label('$U_{BE}$', loc='left')
    d.save('bjt_polarizer.svg', transparent=True)
```

```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing(transparent=True) as d:
    elm.style(elm.STYLE_IEC)
    bjt = elm.BjtNpn(leads=True, circle=True).theta(90).label('Q1')

    # Build base and bias network first so R1 connection exists before we connect Vcc
    base_line = elm.Line().at(bjt.base).down(d.unit/2)
    base_dot = elm.Dot().at(base_line.end)
    elm.Line().at(base_dot.center).down(d.unit/4).dot().label('$U_{BE}$', loc='bottom')

    r1_line = elm.Line().at(base_dot.center).left(d.unit/4)
    R1 = elm.Resistor().at(r1_line.end).left().label('$R_{1}$')
    r1_connect_line = elm.Line().at(R1.end).left(d.unit/4)

    r2_line = elm.Line().at(base_dot.center).right(d.unit/4)
    R2 = elm.Resistor().at(r2_line.end).right().label('$R_{2}$')
    # create a dot at the end of the R2 connection and place a ground there
    r2_connect_dot = elm.Line().at(R2.end).right(d.unit/4).dot()
    elm.Ground().at(r2_connect_dot.end)

    # Collector and Vcc connection (now we can connect to r1_connect_line)
    Rc = elm.Resistor().at(bjt.collector).left().label('$R_{C}$')
    vcc_line = elm.Line().at(Rc.end).left(d.unit/4)
    # join Vcc line to the R1 connection and add a dot there
    elm.Line().at(vcc_line.end).to(r1_connect_line.end).dot()
    # label Ucc downward from the junction
    elm.Line().at(r1_connect_line.end).down(d.unit/4).dot().label('$U_{CC}$', loc='top')

    elm.Line().up(d.unit/4).at(bjt.collector).dot().label('$U_{CE}$', loc='right')
    Re = elm.Resistor().at(bjt.emitter).right().label('$R_{E}$')
    gnd_line = elm.Line().at(Re.end).right(d.unit/4)

    # connect R2 ground to emitter ground line
    elm.Line().at(r2_connect_dot.end).to(gnd_line.end)

    d.save('bjt_polarizer_rotated.svg', transparent=True)

print("Le circuit pivoté a été sauvegardé dans 'bjt_polarizer_rotated.svg'")
```
