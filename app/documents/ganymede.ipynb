```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing(transparent=True) as d:
    elm.style(elm.STYLE_IEC)
    op = elm.Opamp(leads=True)
    elm.Line().down(d.unit/4).at(op.in2)
    elm.Ground(lead=False)
    Rin = elm.Resistor().at(op.in1).left().idot().label('$R_{1}$', loc='bot').dot().label('$v_{e}$', loc='left')
    elm.Line().up(d.unit/2).at(op.in1)
    elm.Resistor().tox(op.out).label('$R_{2}$')
    elm.Line().toy(op.out).dot()
    elm.Line().right(d.unit/4).at(op.out).dot().label('$v_{s}$', loc='right')
    d.save('schematic_transparent.svg', transparent=True)
```

```python
# opampdifferential.py
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing(transparent=True) as d:
    elm.style(elm.STYLE_IEC)
    op = elm.Opamp(leads=True)
    R1 = elm.Resistor().at(op.in1).left().idot().label('$R_{1}$', loc='bot').dot().label('$v_{e}$', loc='left')
    elm.Line().up(d.unit/2).at(op.in1)
    R2 = elm.Resistor().tox(op.out).label('$R_{2}$')
    elm.Line().toy(op.out).dot()
    R3 = elm.Resistor().at(op.in2).left().idot().label('$R_{3}$', loc='bot').dot().label('$v_{e}$', loc='left')
    elm.Line().down(d.unit/2).at(op.in2)
    R4 = elm.Resistor().tox(op.out).label('$R_{4}$')
    elm.Ground().at(R4.end)
    elm.Line().right(d.unit/4).at(op.out).dot().label('$v_{s}$', loc='right')
    d.save('opampdifferential.svg', transparent=True)
```

```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing(transparent=True) as d:
    elm.style(elm.STYLE_IEC)
    x = elm.Transformer(t1=6, t2=5).label('$N_{P}:N_{S}$')
    lines1 = elm.Line().at(x.s1).length(d.unit/1.5).dot()
    elm.CurrentLabelInline(direction='in').at(lines1.end).label('$I_{S}$')
    lines2 = elm.Line().at(x.s2).length(d.unit/1.5).dot()
    elm.Gap().at(lines1.end).to(lines2.end).label(('+','$U_S$','-'))
    linep1 = elm.Line().at(x.p1).length(d.unit/1.5).left().dot()
    elm.CurrentLabelInline(direction='out').at(linep1.end).label('$I_{P}$')
    linep2 = elm.Line().at(x.p2).length(d.unit/1.5).left().dot()
    elm.Gap().at(linep1.end).to(linep2.end).label(('+','$U_P$','-'))
    d.save('transformator.svg', transparent=True)
    
```
